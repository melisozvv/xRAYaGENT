```python
# Basic phrase grounding for chest X-ray findings
from transformers import AutoModelForCausalLM, AutoProcessor
from PIL import Image
import torch
model = AutoModelForCausalLM.from_pretrained("microsoft/maira-2", trust_remote_code=True)
processor = AutoProcessor.from_pretrained("microsoft/maira-2", trust_remote_code=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.eval().to(device)
image = Image.open("../data/xray.jpg")
phrase = "Pleural effusion"
processed_inputs = processor.format_and_preprocess_phrase_grounding_input(
        frontal_image=image,
        phrase=phrase,    
        return_tensors="pt"
        ).to(device)
with torch.no_grad():
    output = model.generate(**processed_inputs, max_new_tokens=150, use_cache=True)
    prompt_length = processed_inputs["input_ids"].shape[-1]
    decoded_text = processor.decode(output[0][prompt_length:], skip_special_tokens=True)
    prediction = processor.convert_output_to_plaintext_or_grounded_sequence(decoded_text)
    print(f"Phrase: {phrase}")
    print(f"Grounding result: {prediction}")

        
```
